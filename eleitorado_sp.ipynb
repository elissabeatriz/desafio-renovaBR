# tratando a base de dados perfil_eleitorado_2020

```python
#importa a biblioteca pandas
import pandas as pd
```

```python
#lê o arquivo csv, com codificação Latin 1 e separado por ';'
df = pd.read_csv("perfil_eleitorado_2020.csv", encoding="latin_1", sep=";")
```

```python
#seleciona as colunas relevantes para análise posterior e remove linhas com valores nulos
df_eleitorado = df[["SG_UF", "CD_MUNICIPIO","NM_MUNICIPIO", "NR_ZONA", "CD_GENERO","DS_GENERO", "CD_ESTADO_CIVIL",
                "DS_ESTADO_CIVIL", "CD_FAIXA_ETARIA", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE", "QT_ELEITORES_PERFIL"]].dropna()
```

```python
#filtra dados apenas para o estado de São Paulo
df_eleitorado = df_eleitorado.query("SG_UF=='SP'")
df_eleitorado
```

```python
#salva os dados tratados em um novo arquivo csv, com codificação Latin 1, separado por ';' e sem índice
df_eleitorado.to_csv("eleitorado_sp.csv", encoding = "latin_1", sep=";", index = False)
```
