# tratando a base de dados SP_turno_1

```python
#Importa a biblioteca pandas
import pandas as pd
```

```python
#Lê o arquivo csv, com codificação Latin 1 e separado por ';'
df = pd.read_csv("SP_turno_1.csv", encoding="latin_1", sep=";")
```

```python
#Seleciona as colunas relevantes para análise posterior e remove linhas com valores nulos
df_votos = df[["NM_VOTAVEL", "DS_CARGO_PERGUNTA", "SG_PARTIDO","NM_PARTIDO", 
               "NM_MUNICIPIO", "CD_MUNICIPIO", "NR_ZONA","NR_SECAO", "QT_VOTOS",
               "NR_URNA_EFETIVADA", "CD_FLASHCARD_URNA_EFETIVADA"]].dropna()
```

```python
#Descarda os votos nulos e brancos
df_votos = df_votos[df_votos["NM_PARTIDO"] != "#NULO#"]
```

```python
#Correções nos nomes de candidatos
df_votos = df_votos.replace({": JUSSARA VIANA": "JUSSARA VIANA", "´SERGIO SOVENIL" : "SERGIO SOVENIL"})
df_votos
```

```python
#Salva os dados tratados em um novo arquivo csv, com codificação Latin 1, separado por ';' e sem índice
df_votos.to_csv("votos.csv", encoding = "latin_1", sep=";", index = False)
```
